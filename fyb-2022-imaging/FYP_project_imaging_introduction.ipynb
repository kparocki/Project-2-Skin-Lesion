# Explore an image from the dataset

```
# Load an image and display it

im = plt.imread('fyp2022-imaging/data/example_image/ISIC_0001769.jpg')
plt.imshow(im)
```

```
#A color image is a array with 3 dimensions (x, y, R-G-B color channels) of integers

print(im.shape)
print(im.dtype)

#Other packages might wrap the image in a different class - you are allowed to use those if you want
```

```

# Get a single RGB value from the blue circle (marker used by dermatologist)
print(im[2000,2000,:])

# Show only the red channel
plt.imshow(im[:,:,0], cmap='gray') #RGB order

```

```
# Display only a part of the image

im_part = im[60:120,130:220,:]
plt.imshow(im_part)

```

```
# Modify the image by setting some pixels to black

im_copy = im_part.copy()


im_copy[0:10,0:10,:] = np.tile(0, [10, 10, 3]) #this basically creates a cube of 0, which is later overlayed over the cube of colors

plt.imshow(im_copy)
```

```
# Show the images overlayed, for this we can use PIL 

#!pip install pillow 
from PIL import Image 

# Load images as Image objects  
img1 = Image.open('fyp2022-imaging/data/example_image/ISIC_0001769.jpg') 
img2 = Image.open('fyp2022-imaging/data/example_segmentation/ISIC_0001769_segmentation.png') 
  
# Overlay - more options such as transparency should be available here  
img2.paste(img1, (0,0), mask = img2) 
  
# Display 
img2.show()  # This doesn't actually display an image in Google Colab :(
plt.imshow(img2, cmap='gray')


# Note that this is a single channel image -> no color data
print(img2.size)          
```

```
# Alternative: replace the non-lesion pixels

img1 = im.copy()
img1[mask==0] = 0 #if the mask is black at some spot, set image color to black at that spot
  
# Display 
plt.imshow(img1)

# You can use any package you prefer, but beware you might need to convert between formats
```

```
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray #RGB taken down to one numerical value

img1 = plt.imread('fyp2022-imaging/data/example_image/ISIC_0001769.jpg')
gray = rgb2gray(img1)

plt.imshow(gray, cmap='gray') 
```

```
#Break down the image RGB channels in the leftmost corner

img1 = plt.imread('fyp2022-imaging/data/example_image/ISIC_0001769.jpg')

r, g, b = img1[0,0,0], img1[0,0,1], img1[0,0,2]
print(r,g,b)
print(img1[0,0,:])
```
