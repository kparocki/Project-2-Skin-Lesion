```python
def middle(mask):
    M = cv2.moments(mask)
    # calculate x,y coordinate of center
    cX = int(M["m10"] / M["m00"])
    cY = int(M["m01"] / M["m00"])
    return cX,cY
```

```python
id_ = df['image_id'][0]
mask = cv2.imread(path.segmentation+f'{id_}_segmentation.png',0)
fig, axs = plt.subplots(ncols=2)
axs[0].imshow(mask)
a,b,c,d = util.extract_bboxes(np.array(mask))
maskCrop = mask[a:b,c:d]
axs[1].imshow(maskCrop)
```

```python
lst = [0,45,90,180,270,55]
def symetryScore(mask):
    
    pivot = middle(mask)
    print(pivot)
    padX = [mask.shape[1] - pivot[0], pivot[0]]
    padY = [mask.shape[0] - pivot[1], pivot[1]]
    imgP = np.pad(mask, [padY, padX], 'constant')
    res = []
    for i,x in enumerate(lst):
        rotated = ndimage.rotate(imgP, x)
        if rotated.shape[1] % 2 == 1:
            rotated = np.pad(rotated, [0,1], 'constant')
        mid = rotated.shape[1]//2
        
        fold = rotated[:,:mid] - np.fliplr(rotated[:,mid:])
        res.append(np.sum(fold))
    
    return np.average(res)

symetryScore(maskCrop)
```
